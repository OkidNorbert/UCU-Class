```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

importing the excel file

```python
# Load the dataset
data = pd.read_excel('Assignment2.xlsx', sheet_name='LifeSavings')
```

Questions
1. Does the dataset have any anomalies such as missing data or outliers? [ 5 MARKS ]

```python
# Check for missing data
print(data.isnull().sum())

#There is no missing data
```

```python
# Detect outliers using Z-scores
z_scores = zscore(data.select_dtypes(include=[float, int]))

outliers = (abs(z_scores) > 3).any(axis=1)

print(f"Outliers:\n{data[outliers]}")
```

2. Calculate the descriptive statistics of all the variables and explain their implication? [ 5 MARKS

```python
# Descriptive statistics for each variable
descriptive_stats = data.describe()
print(descriptive_stats)

#Explanation:
# The output will include mean, min, max, and quartiles,
# helping to interpret the central tendency and variability of each variable.
```

3. Generate a graph showing the distribution of the population under 15 (“pop15”). [ 5 MARKS ]

```python
# Distribution plot for "pop15"
sns.histplot(data['pop15'], kde=True)
plt.title('Distribution of Population Under 15 (pop15)')
plt.xlabel('Percentage of Population Under 15')
plt.ylabel('Frequency')
plt.show()
```

4. What does the relationship between the population over 75 (“pop75”) and the growth rate of
dpi (“ddpi”) look like? Describe the relationship using graphs and statistics. [ 5 MARKS ]

```python
# Scatter plot for "pop75" vs "ddpi"
sns.scatterplot(x='pop75', y='ddpi', data=data)
plt.title('Relationship between Population Over 75 and Growth Rate of DPI')
plt.xlabel('Population Over 75 (pop75)')
plt.ylabel('Growth Rate of DPI (ddpi)')
plt.show()

# Calculate correlation
correlation = data['pop75'].corr(data['ddpi'])
print(f'Correlation between pop75 and ddpi: {correlation}')
```

5. What is the relationship between aggregate personal saving (“sr”) and real rear disposal
income (“dpi)”? [ 5 MARKS ]

```python
# Scatter plot for "sr" vs "dpi"
sns.scatterplot(x='dpi', y='sr', data=data)
plt.title('Relationship between Aggregate Personal Saving (sr) and Real Disposable Income (dpi)')
plt.xlabel('Real Disposable Income (dpi)')
plt.ylabel('Aggregate Personal Saving (sr)')
plt.show()

# Calculate correlation
correlation_sr_dpi = data['sr'].corr(data['dpi'])
print(f'Correlation between sr and dpi: {correlation_sr_dpi}')
```
